#### Penskalaan Fitur

Pada tahap ini, kita melakukan penskalaan fitur atau normalisasi yaitu mengubah nilai pada fitur agar lebih mudah untuk diolah oleh sistem dan dengan hasil yang lebih maksimal. Setelah itu, kita akan melakukan proses feature extraction dengan teknik PCA (Principal Component Analysis). Yaitu proses untuk menggabungkan beberapa fitur menjadi 1 atau beberapa fitur. Tujuannya, untuk mengurangi proses komputasi yang dilakukan. Karena semakin banyak fitur yang diolah, semakin lama juga proses komputasi akan berjalan.

```python
cd_features = car_data_v1.drop(['DateCrawled', 'DateCreated', 'LastSeen', 'Price'], axis=1)
scaler = StandardScaler()
scaler.fit(cd_features[['RegistrationYear', 'RegistrationMonth', 'Power', 'Mileage', 'NumberOfPictures', 'PostalCode']])
cd_features.loc[:, ['RegistrationYear', 'RegistrationMonth', 'Power', 'Mileage', 'NumberOfPictures', 'PostalCode']] = scaler.transform(cd_features[['RegistrationYear', 'RegistrationMonth', 'Power', 'Mileage', 'NumberOfPictures', 'PostalCode']])
cd_features = pd.get_dummies(cd_features, columns=categorical_features)
cd_features
```

```python
cd_target = car_data_v1['Price']
cd_target
```

```python
pca = PCA(n_components=1)

data_pca = pca.fit_transform(cd_features)
```

```python
cd_features_v1 = pd.DataFrame(data_pca, columns=['feature1'])
cd_features_v1
```

```python
cd_target
```

#### Pemisahan Data Uji

```python
X_train, X_valid, y_train, y_valid = train_test_split(cd_features_v1, cd_target, test_size=0.4, random_state=12345)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size=0.5, random_state=12345)
```

## Pelatihan Model

Setelah proses persiapan data selesai, sekarang saatnya proses pelatihan model dengan metode Linear Regression, Decision Tree, Random Forest, dan metode gradient boosting seperti LightGBM, CatBoost, dan XGBoost. Untuk mengukur kualitas dari model, kita akan menggunakan nilai RMSE (Root Mean Square Error). Oleh sebab itu, pertama kita perlu membuat sebuah fungsi RMSE untuk mempermudah proses perhitungan kualitas model pada saat pelatihan.

```python
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred) ** 0.5
```

### Regresi Linear

```python
%%time
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_predict = lr.predict(X_valid)
print(f'RMSE for Linear Regression: {rmse(y_valid, lr_predict)}')
```

### Decision Tree

```python
%%time
dt = None
dt_score = 99999
dt_hyperparam = None
for depth in range(1,11):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    result = rmse(y_valid, predictions)
    if result < dt_score:
        dt = model
        dt_score = result
        dt_hyperparam = {
            "depth": depth
        }
        #print(f'depth: {depth}, RMSE: {result}')        

print(f'Best depth: {dt_hyperparam["depth"]}, score: {dt_score}')
```

### Random Forest

```python
%%time
rf = None
rf_hyperparam = None
rf_score = 99999
for est in range (10,51,10):
    for depth in range(1,11):
        model = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth)
        model.fit(X_train, y_train)
        predictions_valid = model.predict(X_valid)
        result = (rmse(y_valid, predictions_valid))
        if(result < rf_score):
            rf_score = result
            rf_hyperparam = {
                "est": est,
                "depth": depth
            }            
            rf = model
        #print(f'RMSE: {result}, est: {est}, max depth: {depth}')

print(f'{rf_hyperparam}, score: {rf_score}')        
```

### Gradient Boosting - LightGBM

```python
%%time
lgbm = None
lgbm_hyperparam = None
lgbm_score = 99999
for est in range (10,51,10):
    for depth in range(1,11):
        for alpha in range(1,11):
            model = LGBMRegressor(random_state=12345, n_estimators=est, max_depth=depth, reg_alpha=alpha)
            model.fit(X_train, y_train)
            predictions_valid = model.predict(X_valid)
            result = (rmse(y_valid, predictions_valid))
            if(result < lgbm_score):
                lgbm_score = result
                lgbm_hyperparam = {
                    "est": est,
                    "depth": depth,
                    "alpha": alpha
                }            
                lgbm = model
            #print(f'RMSE: {result}, est: {est}, max depth: {depth}, alpha: {alpha}')

print(f'{lgbm_hyperparam}, score: {lgbm_score}') 
```

### Gradient Boosting - CatBoost

```python
%%time
cb = None
cb_hyperparam = None
cb_score = 99999
for depth in range(1,11):
    model = CatBoostRegressor(random_state=12345, iterations=150, depth=depth)
    model.fit(X_train, y_train)
    predictions_valid = model.predict(X_valid)
    result = (rmse(y_valid, predictions_valid))
    if(result < cb_score):
        cb_score = result
        cb_hyperparam = {
            "depth": depth,
        }            
        cb = model

print(f'{cb_hyperparam}, score: {cb_score}') 
```

### Gradient Boosting - XGBoost

```python
%%time
xgb = None
xgb_hyperparam = None
xgb_score = 99999

for depth in range(1,11):
    model = XGBRegressor(random_state=12345, max_depth=depth, reg_alpha=1)
    model.fit(X_train, y_train)
    predictions_valid = model.predict(X_valid)
    result = (rmse(y_valid, predictions_valid))
    if(result < xgb_score):
        xgb_score = result
        xgb_hyperparam = {
            "est": est,
            "depth": depth
        }            
        xgb = model
        #print(f'RMSE: {result}, est: {est}, max depth: {depth}')

print(f'{xgb_hyperparam}, score: {xgb_score}')
```

Proses pelatihan model diperkirakan membutuhkan waktu selama 10 menit. Dengan proses pelatihan paling pada LinearRegression, karena memang tidak ada proses iteratif pada pelatihan Linear Regression. Sedangkan LightGBM membutuhkan waktu paling lama dikarenakan memang proses iteratif yang bercabang hingga 3x. Untuk model XGBoost juga bisa disebut paling lama karena proses pelatihan hanya dengan 1x iteratif tapi membutuhkan waktu hingga hampir 3 menit. Dan dari hasil RMSE, model Random Forest memiliki hasil paling kecil dibanding model lain. Berikutnya, kita akan melakukan analisa dari model terkait kualitas, dan kecepatan prosesnya.
