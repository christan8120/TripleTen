Sekarang, kita akan mengecek berapa banyak data yang hilang di setiap kolom. Kita akan memutuskan apakah baris dengan data yang hilang perlu kita hapus dari daftar analisa data atau kita gantikan nilai yang hilang tersebut dengan data yang lebih sesuai.

```python
def create_null_report(df):
    null_counts = df.isnull().sum().to_frame().rename(columns={0: 'missing_values'})
    null_counts['%_of_total'] = (null_counts['missing_values']/len(df)).round(2) *100
    null_table = null_counts.sort_values(by='missing_values', ascending=False)
    return null_table
```

Lalu, untuk kolom year_of_release tidak ada pola khusus yang terlihat menyebabkan kolom tersebut berisi nilai yang hilang. Sedangkan untuk kolom genre dan name memiliki pola yang sama, dan juga kolom lainnya seperti critic_score, user_score dan rating juga kosong. 

Pertama, kita perlu memperbaiki nilai yang hilang pada kolom kategorikal, mulai dari kolom name, dan genre dimana kita bisa menghapus nilai yang hilang pada 2 kolom ini. Sedangkan untuk nilai hilang pada kolom year_of_release, karena hanya ada maksimal 2% data dari total baris maka nilai yang hilang pada kolom ini diubah menjadi 0. Sedangkan untuk kolom rating, kita perlu mengubah nilai yang hilang dengan value yang lebih baik. Disini penulis akan mengisi nilai yang kosong pada kolom rating dengan value "Unknown".

```python
df = df.dropna(subset=['genre', 'name'])
df['year_of_release'] = df['year_of_release'].fillna(0)
#df = df[df['user_score'] != 'tbd']
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
df['rating'] = df['rating'].fillna('Unknown')
create_null_report(df)
```

Sekarang, tersisa 2 kolom dengan nilai yang hilang yaitu critic_score dan user_score. Kedua kolom ini memiliki persentase data yang hilang sangat banyak, mencapai separuh dari total data yang kita miliki. Sebelum kita mengisi nilai yang hilang pada kedua kolom ini, kita perlu memperbaiki tipe data pada kolom year_of_release dan user_score terlebih dahulu. Karena kolom ini nanti akan berguna untuk menentukan value apa yang perlu diisi ke 2 kolom dengan nilai yang hilang.

```python
df['year_of_release'] = df['year_of_release'].astype('int')
df['user_score'] = df['user_score'].astype('float')
```

```python
df.describe()
```

Dengan informasi diatas, kita akan mengganti nilai yang kosong pada kolom critic_score dan user_score dengan nilai median bedasarkan kategori platform dan genre. Kita menggunakan nilai median untuk menghindari kemungkinan adanya outlier yang dapat mempengaruhi hasil rata-rata atau mean. Disini, kita perlu memastikan apakah nilai median pada kategori platform dan genre benar ada atau justru semuanya berisi nilai yang hilang. Oleh sebab itu, penulis membuat 2 kategori yaitu median berdasarkan platform dan genre bila terdapat nilai yang memiliki isi, dan median berdasarkan genre.

```python
df_na_median_platform_genre = df.pivot_table(index=['platform', 'genre'], values=['critic_score', 'user_score'], aggfunc='median')
df_na_median_genre = df.pivot_table(index=['genre'], values=['critic_score', 'user_score'], aggfunc='median')

for p in df['platform'].unique():
    for g in df[df['platform'] == p]['genre'].unique():        
        #df_na_median['critic_score'][p][g]
        if ({(p, g)}.issubset(df_na_median_platform_genre.index)) == False:
            df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['critic_score'].isna()), 'critic_score'] = df_na_median_genre['critic_score'][g]
            df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['user_score'].isna()), 'user_score'] = df_na_median_genre['user_score'][g]
        else:
            if pd.isna(df_na_median_platform_genre['critic_score'][p][g]) == False:
                df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['critic_score'].isna()), 'critic_score'] = df_na_median_platform_genre['critic_score'][p][g]
            else:
                df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['critic_score'].isna()), 'critic_score'] = df_na_median_genre['critic_score'][g]    
            if pd.isna(df_na_median_platform_genre['critic_score'][p][g]) == False:
                df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['user_score'].isna()), 'user_score'] = df_na_median_platform_genre['user_score'][p][g]
            else:
                df.loc[(df['platform'] == p) & (df['genre'] == g) & (df['user_score'].isna()), 'user_score'] = df_na_median_genre['user_score'][g]
                                
create_null_report(df)
```

Sekarang, semua nilai yang hilang telah teratasi, dan tidak ada lagi masalah pada data yang kita miliki. Semua data telah diproses dan diperbaiki.
```python
#Uji hipotesis
alpha = 0.05

df_above_2005_xbox_one = df_above_2005.query('platform == "XOne"')
df_above_2005_pc = df_above_2005.query('platform == "PC"')

result_1 = st.ttest_ind(df_above_2005_xbox_one['user_score'], df_above_2005_pc['user_score'], equal_var=False)

if result_1.pvalue < alpha:
    print('Kita menolak hipotesis nol. Rata-rata rating pengguna untuk Xbox One dan PC berbeda.')
else:
    print('Kita menerima hipotesis nol. Rata-rata rating pengguna untuk Xbox One dan PC sama.')
```

```python
#Uji hipotesis
alpha = 0.05

df_above_2005_action = df_above_2005.query('genre == "Action"')
df_above_2005_sports = df_above_2005.query('genre == "Sports"')

result_2 = st.ttest_ind(df_above_2005_action['user_score'], df_above_2005_sports['user_score'], equal_var=False)

if result_2.pvalue < alpha:
    print('Kita menerima hipotesis nol. Rata-rata rating pengguna untuk genre Action dan Sports berbeda.')
else:
    print('Kita menolak hipotesis nol. Rata-rata rating pengguna untuk genre Action dan Sports sama.')
```
