 Tidak ada data yang terduplikasi, sehingga proses penghapusan data yang terduplikasi bisa diabaikan.

```python
def create_null_report(df):
    null_counts = df.isnull().sum().to_frame().rename(columns={0: 'missing_values'})
    null_counts['%_of_total'] = (null_counts['missing_values']/len(df)).round(2) *100
    null_table = null_counts.sort_values(by='missing_values', ascending=False)
    return null_table
```

## Mempelajari dan Menangani Outlier


```python
def upper_lower_whisker(series):
    Q1  = series.quantile(0.25)
    Q3  = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - (1.5 * IQR)
    upper_whisker = Q3 + (1.5 * IQR)
    return lower_whisker, upper_whisker
```

```python
def outlier_cleansing(dataframe, column):

    lower, upper = upper_lower_whisker(dataframe[column])
    print(f'lower values : {lower}')
    print(f'upper values : {upper}')

    plt.title('Before cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    dataframe = dataframe[(dataframe[column] > lower) & (dataframe[column] < upper)]

    lower_check, upper_check = upper_lower_whisker(dataframe[column])
    print(f'lower_check values : {lower_check}')
    print(f'upper_check values : {upper_check}')

    plt.title('After cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    if lower == lower_check or upper == upper_check:
        print(f'\n====== [FINISHED] outlier in [{column}] column has been clear! ====== \n')
        return dataframe

    else:
        print(f'\n------ [RE-START] cleaning outlier in {column} column ------ \n')
        dataframe = outlier_cleansing(dataframe, column)
        return dataframe
```

```python
df_final = df.copy()
list_of_selected_column = ['condition_level', 'price', 'odometer_year', 'vehicle_age', 'cylinders']

for column in list_of_selected_column:
    print(f'\n====== [START] cleaning outlier in {column} column ====== \n')    
    df_final = outlier_cleansing(dataframe=df_final, column=column)
```

Proses penanganan outlier terlihat kompleks. Sederhananya, ada 3 proses utama yang kita lakukan:
1. Menentukan batas outlier atas dan bawah
2. Memilih dan mengeluarkan data yang melebihi batas outlier
3. Membuat sebuah dataframe baru dengan data yang yang lebih baik (tidak ada outlier)

Mari kita pahami satu per satu prosesnya. Pertama, kita membuat sebuah fungsi yang akan mengembalikan data outlier atas dan bawah dari sebuah series. Fungsi ini dibuat untuk mempermudah kita dalam menentukan outlier yang ada dimasing-masing kolom, sehingga menghindari code redundancy. Untuk menggunakan fungsi ini, kita bisa memanggil upper_lower_whisker(df[nama_kolom]), dimana df[nama_kolom] disini berbentuk Series.

Proses berikutnya adalah membuat sebuah fungsi untuk mengeluarkan data-data yang melebihi batas outlier. Ada 3 proses pada fungsi ini, pertama kita tentukan terlebih dahulu outlier di kolom yang sudah ditentukan dari dataframe (dataframe disini bisa disebut dataframe pertama). Batas outlier bisa kita dapatkan dengan memanggil function upper_lower_whisker yang sudah kita buat sebelumnya. Setelah itu kita filter dataframe pertama dengan mengambil data yang tidak keluar dari outlier (bisa kita sebut dataframe kedua karena sudah difilter), terakhir kita bandingkan 2 outlier dari 2 dataframe ini. Bila, 2 outlier ini hasilnya sama, maka proses perbaikan bisa dianggap selesai, sebaliknya bila hasil dari 2 outlier ini berbeda, maka proses akan kembali diulang. Perulangan disini dilakukan dengan memanggil function itu sendiri, atau bisa kita sebut sebagai recursive. dataframe=outlier_cleansing(...) adalah contoh recursive. Proses recursive ini akan terus dilakukan hingga 2 batas outlier dari 2 dataframe memiliki hasil yang sama.

Proses terakhir adalah membuat sebuah variable yang berisi data yang sudah diexclude dari outlier yang bisa mempengaruhi hasil analisa. Disini kita membuat 2 variable yaitu df_final dan columns. Hasil dataframe yang sudah diperbaiki disimpan dalam variable df_final, sedangkan columns akan berisi daftar kolom yang perlu kita cek outlier-nya. Untuk mendapatkan hasil dataframe yang sudah final, kita hanya perlu memanggil function outlier_cleansing yang sudah dibuat.

## Kesimpulan Umum

Kita mendapatkan data yang perlu kita analisa untuk bisa menentukan apa saja faktor yang mempengaruhi harga kendaraan, khususnya terkait dengan kondisi, jarak tempuh, umur kendaraan, transmission dan warna. Dari data yang kita terima, ada beberapa masalah yang perlu diperbaiki sebelum data dapat diproses. Masalah tersebut seperti nilai yang hilang dan kesalahan tipe data. Masalah ini dapat kita atasi dengan melihat behaviour pada nilai yang hilang, dan value yang ada. Contohnya untuk beberapa data numeric dapat kita ubah nilai yang hilang dengan nilai median. Sedangkan untuk kolom dengan tipe data boolean dapat kita ubah nilai yang hilang menjadi False. 

Setelah proses perbaikan data selesai, kita perlu menentukan batas upper dan lower outlier, dan melakukan filtering terhadap dataframe. Pada tahap ini, kita melakukan pengecekan outlier pada 3 parameter numerik inti, yaitu kondisi, odometer dan umur kendaraan. Proses ini diperlukan karena data outlier dapat mempengaruhi hasil analisa, dan juga ini dilakukan untuk memberikan hasil akhir analisa atau kesimpulan yang tepat.

Setelah data dengan kualitas yang bagus kita dapatkan, barulah kita mulai proses analisa relasi antara parameter inti dengan harga. Disini, kita mengambil 2 jenis kendaraan dengan total penayangan iklan yang paling banyak, yaitu sedadn dan SUV. Dari hasil tersebut kita dapat melihat bahwa korelasi antara kondisi dan harga untuk 2 jenis kendaraan cenderung meningkat namun tidak terlalu kuat. Yang artinya semakin bagus kondisi kendaraan bisa saja mempengaruhi harganya. Sebaliknya, korelasi antara jarak tempuh atau odometer dan umur kendaraan dengan harga berhasil minus dan cukup kuat, yang artinya semakin tinggi odometer dan umur kendaraan, maka semakin murah juga harga kendaraan. 

Selain data numerik, data kategorikal yaitu jenis transmission untuk kendaraan SUV juga mempengaruhi harga kendaraan, yaitu untuk jenis transmission manual memiliki rata-rata lebih tinggi ketimbang automatic. Sebaliknya untuk kendaraan sedan, harga sedan dengan transmisi automatic sedikit lebih tinggi ketimbang manual. Terakhir, warna kendaraan untuk jenis sedan sepertinya tidak mempengaruhi harga kendaraan, karena rata-rata harga cenderung sama. Tetapi untuk jenis SUV, warna orange memiliki harga paling tinggi ketimbang warna lainnya, namun harga tertinggi cenderung sama dengan warna lainnya. Kesimpulannya, dari 5 paramater yang sebelumnya kita anggap mempengaruhi harga, 4 diantaranya bisa kita simpulkan benar, yaitu kondisi, odometer, umur kendaraan dan transmission. Tetapi untuk paramter terakhir yaitu warna, mungkin ini bisa mempengaruhi harga di mobil SUV, namun tidak mempengaruhi harga di mobil sedan. Kita bisa sederhanakan bahwa bila mobil dengan harga yang tinggi, bisa saja karena kondisi yang masih bagus, odometer yang masih sedikit, umur kendaraan yang belum terlalu lama, dan jenis transmisi yang digunakan.
